# torch vision dataasets

```python
import torch
import matplotlib.pyplot as plt
import numpy as np
import torchvision.transforms as transforms
import torchvision.datasets as dsets
```

```python
torch.manual_seed(0)
```

```python
def show_data(data_sample, shape=(28,28)):
    plt.imshow(data_sample[0].numpy().reshape(shape),cmap='gray')
    plt.title('y='+str(data_sample[1]))
```

```python
dataset=dsets.MNIST(
    root='datasets/',
    download=True,
    transform=transforms.ToTensor()
)
```

```python
show_data(dataset[0])
```

```python
show_data(dataset[1])
```

# torch vision transform

```python
croptensor_transform=transforms.Compose([transforms.CenterCrop(20),transforms.ToTensor()])
dataset=dsets.MNIST(root='datasets/',download=False,transform=croptensor_transform )
dataset[0][0].shape
```

```python
show_data(dataset[2],shape=(20,20))
```

```python
fliptensor_data_transform = transforms.Compose([transforms.RandomHorizontalFlip(p = 1),transforms.ToTensor()])
dataset = dsets.MNIST(root = './data', download = True, transform = fliptensor_data_transform)
show_data(dataset[1])
```

```python
all_transforms=transforms.Compose([transforms.RandomVerticalFlip(p=1),transforms.RandomHorizontalFlip(p=1),transforms.ToTensor()])
dataset=dsets.MNIST(root='datasets/',download=False,transform=all_transforms)
dataset[0][0].shape
```

```python
show_data(dataset[5])
```
