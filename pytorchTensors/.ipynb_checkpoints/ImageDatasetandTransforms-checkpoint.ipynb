# create a dataset class
- dataset class
- transform  

```python
class Dataset(Dataset):
    def __init__(self,csv_file,data_dir,transform=None):
        self.data_dir=data_dir
        self.transform=transform
        data_dircsv_file=os.path.join(self.data_dir,csv_file)
        self.data_name=pd.read_csv(data_dircsv_file)
        self.len=self.data_name.shape[0]

    def __len__(self):
        return self.len

    def __getitem__(self,idx):
        img_name=os.path.join(self.data_dir,self.data_name.iloc[idx,1])
        image=Image.open(img_name)
        y=self.data_name.iloc[idx,0]

        if self.transform:
            image=self.transform(image)
        return image,y
```

# torch vision transforms 

```python
import torchvision.transforms as transforms 
```

```python
croptensor_data_trasform= transforms.Compose([transforms.CenterCrop(20),transforms.ToTensor()])
dataset=Dataset(csv_file=csv_file,data_dir=directory,transform=croptensor_data_trasform)
dataset[0][0].shape
```

```python
show_data(dataset[0],shape=(20,20))
```

```python
show_data(dataset[1],shape=(20,20))
```

```python
fliptensor_data_transform= transforms.Compose([transforms.RandomVerticalFlip(p=1),transforms.ToTensor()])
dataset=Dataset(csv_file=csv_file, data_dir=directory, transform=fliptensor_data_transform)
```

```python
# dataset[0][0].shape
show_data(dataset[11],shape=(28,28))
```

```python
transform_vertical_flip=transforms.Compose([transforms.RandomVerticalFlip(p=1),transforms.ToTensor()])
dataset=Dataset(csv_file=csv_file, data_dir=directory, transform=transform_vertical_flip)
dataset[0][0].shape
```

```python
show_data(dataset[10])
```
